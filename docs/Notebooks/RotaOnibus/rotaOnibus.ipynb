```python
import networkx as nx
import osmnx as ox
import folium

bus_stops = [
    {'id': 0, 'lat': -15.998730, 'lon': -48.057907},
    {'id': 1, 'lat': -15.793767, 'lon': -47.882389},
]

# Carregue a rede de ruas do arquivo .graphml
G_DF = ox.load_graphml("private_false_distrito_federal.graphml")

# Encontre os nós mais próximos para os pontos
nearest_nodes = [ox.distance.nearest_nodes(G_DF, stop['lon'], stop['lat']) for stop in bus_stops]

# Calcule a rota entre todos os pontos
all_routes = []
for i in range(len(nearest_nodes) - 1):
    route_segment = nx.shortest_path(G_DF, nearest_nodes[i], nearest_nodes[i + 1], weight='length')
    all_routes.extend(route_segment[:-1])  # Não adicione o último ponto, pois ele será o primeiro do próximo segmento
all_routes.append(nearest_nodes[-1])  # Adicione o último ponto

# Plote a rota usando folium
m = folium.Map(location=[bus_stops[0]['lat'], bus_stops[0]['lon']], zoom_start=10)
route_coords = [(G_DF.nodes[node]['y'], G_DF.nodes[node]['x']) for node in all_routes]
folium.PolyLine(route_coords, color="blue", weight=2.5, opacity=1).add_to(m)

for stop in bus_stops:
    folium.Marker([stop['lat'], stop['lon']], popup=str(stop['id'])).add_to(m)

m.save("route_map.html")
m
```
